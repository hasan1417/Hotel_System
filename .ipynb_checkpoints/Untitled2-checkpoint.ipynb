# Introduction:

# Methodology:

#### First of all, two important built-in python libraries were used. First, dateime library which was used to deal with dates. Second, random library which was used in order to generate random numbers for the confirmation number.

```python
# Import two libraries
import datetime
import random
```

#### The first function was made is "start_system" which will prompt the user for the main menu of the hotel system. First, it will show the user the available options and it will integrate all the functions together and call them corresponding to the entered option. Furthermore, the function will ensure all the options are valid and will lead to a successful call of the needed option. below is screenshot for the working "start_system" function:

```python
#Make new function to act as interface for the user
def start_system():
    #Printing the avaliable options for the user & Prompting the user to enter a choice
    choice = input("""
    \t\t\t\t\t\t Hotel Booking System\n
    \t\t\t 1 Add New Hotel Booking\n
    \t\t\t 2 Generate Booking Confirmation\n
    \t\t\t 3 Display Information for all Bookings\n
    \t\t\t 4 Display Specific Booking\n
    \t\t\t 5 Update Booking Information\n
    \t\t\t 6 Exit\n
    Enter your choice:""")
    
    #Making sure the user inputs the right value using while loop
    options = [option for option in range(1,7)]
    while int(choice) not in options: 
        choice = input("Please input a correct option: ")
        
    #using match statement to direct the user to right function corresponding to the option was chosen
    match int(choice):
        #if option 1 was chosen the user will be asked to create a new booking
        case 1:
            create_booking()
            
        #if option 2 was chosen the user will generate the confirmation_lastname.txt file          
        case 2:
            generate_booking()
            
        #if option 3 was chosen all the bookings will be displayed in nice format            
        case 3:
            display_all_booking()
            
        #if option 4 was chosen the user will be asked to which option to check his booking               
        case 4:
            #ask the user to pick 1 for using confirmation number or 2 for phone number
            print("\t\t\t choose 1 to search by confirmation number\n")
            print("\t\t\t choose 2 to search by phone number\n")
            #while loop to make sure the user inputted the right option
            while True:                
                option = input("Enter option: ")
                if option == '1' or option == '2':
                    break
                print('Choose Valid Option!')

            #match statement is used to make sure the user has enter a right confirmation/phonen number
            match option: 
                #if the user picked 1 for confirmation number
                case '1':
                    #while loop to validate a right format confirmation number was used
                    while True:
                        confirmation_number = input('Enter your confirmation number: ')
                        #making sure the confirmation number entered is 6 digits and starts with two letters
                        if confirmation_number[:2].isalpha and confirmation_number[2:].strip().isdigit() and len(confirmation_number) == 6:
                            break
                        print('Invalid confirmation number format')
                    #call the function to display the booking information based on the provided confirmation number    
                    display_specific_booking(confirmation_number.upper())
                    
                #if the user picked 2 for phone number                    
                case '2':
                    #while loop to validate a right format phone number was used
                    while True:
                        phone_number = input('Enter your phone number: ')
                        #making sure the enter value is all numbers
                        if phone_number.strip().isdigit():
                            #making sure the value entered is 10 digit and starts with 05
                            if len(phone_number) == 10 and phone_number[:2] == '05':
                                break
                        print('Enter a valid phone number')
                    #call the function to display the booking information based on the provided phone number                           
                    display_specific_booking(phone_number)
                    
        #if option 5 was chosen the user will be asked with three options; update or delete booking or return to menu             
        case 5:
            #print all the valid options for the user
            print("\t\t\t 1 Update Booking information\n")
            print("\t\t\t 2 Cancel Hotel Booking information\n")
            print("\t\t\t 3 Back to the main menu\n")
            #while loop to make the sure has inputted a correct option
            while True:                
                option = input("Enter option: ")
                if option == '1' or option == '2' or option =='3':
                    break
                print('Choose Valid Option!')
            print(option)
            #call the update_booking function with the chosen option
            update_booking(option)

        #if option 6 was chosen the program will be terminated            
        case 6:
            exit()
```

#### The second function is "create_booking" function,  this function will ask the user muliple questions about the followings:
#### - First name
#### - Last name
#### - Phone number
#### - City name
#### - Hotel name
#### - Check-in date
#### - Check-out date
#### - Number of guests
#### in each question was asked, the function will make sure all the entered information are correct and has right format. After all the information are collected, it will be saved in a file called booking.txt

```python
#make new function to help the user to create a successful booking
#in this function will ask ther user to provide his information for the booking and at the end,
#his information will be added to booking.txt
def create_booking():
    
    #first we will ask the user to enter his first name
    #while loop is used to make sure the entered name is all letters
    while True:
        first_name=input("What is your first name? ").lower()
        if first_name.isalpha():
            break
        print ("invalid name")
        
    #second we will ask the user to enter his last name
    #while loop is used to make sure the entered name is all letters
    while True:
        last_name=input("What is your last name? ").lower()
        if last_name.isalpha():
            break
        print ("invalid name")
  
    #then we will ask the user to enter his phone number
    #while loop is used to make sure the entered phone number has correct format
    while True:
        phone_number=input("What is your phone number? ")
        #making sure the enter value is all numbers       
        if phone_number.strip().isdigit():
            #making sure the value entered is 10 digit and starts with 05
            if len(phone_number) == 10 and phone_number[:2] == '05':
                break
        print ("invalid phone number")

    #next we will ask the user to enter his city name
    #while loop is used to make sure the entered name is all letters
    while True:
        city=input("What is your city? ").lower()
        if city.isalpha():
            break
        print ("invalid city name")
 
    #next we will ask the user to enter his hotel name
    #while loop is used to make sure the entered name is all letters
    while True:
        hotel=input("What is the hotel name? ").lower()
        if hotel.isalpha():
            break
        print ("invalid hotel name")
        
    #next we will ask the user to enter his check in date
    #while loop is used to make sure the enter date is in correct format and appropriate 
    while True:
        check_in = input('Enter your check in date: ')
        #try statement to make sure the user enter the date in correct format
        try:
            check_in = datetime.datetime.strptime(check_in, '%d/%m/%Y')
            present = datetime.datetime.now()
            #making sure the check in date is not in the past
            if check_in.date() >= present.date():
                break
            else:
                print('you check in date is in the past!')
        #except statement to alert the user when he entered a wrong format date       
        except:
            print('please enter the date in day/month/year format')
 
    #next we will ask the user to enter his check out date
    #while loop is used to make sure the enter date is in correct format and appropriate 
    while True:
        check_out = input('Enter your check out date: ')
        
        #try statement to make sure the user enter the date in correct format        
        try:
            check_out = datetime.datetime.strptime(check_out, '%d/%m/%Y')
            #making sure the check out date is after the check in date
            if check_out.date() > check_in.date():
                break
            else:
                print('your check out date should be after the check in date!')
        except:
            print('please enter the date in day/month/year format')
            
    #finally, we will ask the user to enter the number of guests
    #while loop is used to make sure the entered value is a number        
    while True:
        guests_number=input("Number of people staying: ")
        #making sure the value entered is number
        if guests_number.strip().isdigit():
            break
        print ("please enter a valid number!")
        
    #generating the confirmation number    
    confirmation_NO = first_name[0].upper() + last_name[0].upper() + f'{random.randint(1000,9999)}'  
    
    #combining the first and last name and assign them to full_name variable
    full_name = first_name + " " + last_name
    
    #listing all the user information preparing it to be fetched to the booking.txt file
    booking_info = [confirmation_NO, full_name, phone_number, city, hotel, check_in.strftime("%d/%m/%Y"), check_out.strftime("%d/%m/%Y"), guests_number]
        
    #here we will create a booking.txt file or append it if it exists.    
    with open('booking.txt','a+') as booking:
        #here we will add our list of the user information to the file
        booking.writelines(','.join(booking_info))
        booking.write('\n')
    print("The booking has been created successfully!")
```

#### The third function used in this program is "generate_booking" function, this function first will make sure that booking.txt files exits. Then if the files exists, it will parse all the information the file and format them nicely. After, formatting them, a confirmation_aleiydi.txt file will be generated and fetch the formatted information inside it.

```python

def generate_booking():
    try:
        with open('booking.txt','r') as booking:
            booking_info = booking.readlines()

        with open('confirmation_aleiydi.txt', 'w') as booking_confirmation:
            booking_confirmation.write("{:<15}    {:<15}   {:<10}   {:<8} {:<10} {:<10} {:<10} {:<2} ".format('Confirmation NO','Full Name','Phone','City','Hotel','Check-in','Check-out','Guests NO'))
            booking_confirmation.write('\n')
            for each_booking in booking_info:
                booking_confirmation.write("{:<15}    {:<15}   {:<10}   {:<8} {:<10}{:<10}   {:<10}     {}".format(*each_booking.split(',')))
        print('The booking confirmation has been generated successfully !')
                
    except:
        print('booking.txt was not found!')
```

```python
def display_all_booking():
    
    try:
        with open('confirmation_lastname.txt','r') as booking_confirmation:
            for line in booking_confirmation.readlines():
                print(line)
    except:
        print('Generate the booking confirmation file please')
        
```

```python
def display_specific_booking(field):
    with open('booking.txt') as booking:
        booking_info = ['Confirmation Number: ','Full Name: ', 'Phone Number: ','City: ','Hotel: ','Check-in: ','Check-out: ','Number of Guests: ']
        all_booking = booking.readlines()
        booking_found = ''
        for each_booking in all_booking:
            if field in each_booking.split(','):
                booking_found = each_booking.split(',')
        if booking_found == '':
            print('Your Entered input did not match any record!')
            start_system()
        else:
            for index, booking_detail in enumerate(booking_found):
                print(booking_info[index], booking_detail)
    
```

```python
def update_booking(choice):
    if choice == '3':
        start_system()
        
    generate_booking()   
    display_all_booking()
    correct_input = False
    while  correct_input == False:
        confirmation_number = input('Enter your confirmation number or 1 to return: ')
        confirmation_number= confirmation_number.upper()
        if confirmation_number[:2].isalpha and confirmation_number[2:].strip().isdigit() and len(confirmation_number) == 6:
            with open('booking.txt','r') as booking_confirmation:
                for booking in booking_confirmation.readlines():
                    if confirmation_number in booking.split(','):
                        correct_input = True
                if correct_input == False:
                    print('Please type a correct confirmation number')
        if confirmation_number == '1':
                correct_input = True
                start_system()
    
    match choice:
        case '1':
            print("\t\t\t 1 Change the hotel name\n")
            print("\t\t\t 2 Change the Check in date\n")
            print("\t\t\t 3 Change the Check out date\n")
            while True:
                option = input('Please enter the number: ')
                if option == '1' or option == '2' or option =='3':
                    break
                print('Choose Valid Option!')
            
            match option:
                
                case '1':
                    while True:
                        hotel=input("What is the hotel name? ").lower()
                        if hotel.isalpha():
                            break
                        print ("invalid hotel name")
                        
                    with open('booking.txt','r+') as all_booking:
                        x=[]
                        for booking in all_booking.readlines():
                            if confirmation_number in booking.split(','):
                                booking_new = booking.split(',')
                                booking_new[-4] = hotel
                                x.append(','.join(booking_new))
                            else:
                                x.append(booking)
                        all_booking.truncate(0)
                        
                    with open('booking.txt','w') as all_booking:
                        for i in x:
                            all_booking.writelines(i)
                
                case '2':
                    while True:
                        check_in = input('Enter your check in date: ')

                        try:
                            check_in = datetime.datetime.strptime(check_in, '%d/%m/%Y')
                            with open('booking.txt', 'r') as booking:
                                all_booking = booking.readlines()
                                for booking in all_booking:
                                    if confirmation_number in booking.split(','):
                                        booking_new = booking.split(',')
                            check_out = datetime.datetime.strptime(booking_new[-2], '%d/%m/%Y')
                            present = datetime.datetime.now()
                            
                            if check_in.date() < present.date():
                                print('You check in should not be in the past!')
                            elif check_in.date() > check_out.date():
                                print('your check in date should be before the check out date!')
                            else:
                                break
                                
                        except:
                            print('please enter the date in day/month/year format')
                            
                    with open('booking.txt','r+') as all_booking:
                        x=[]
                        for booking in all_booking.readlines():
                            if confirmation_number in booking.split(','):
                                booking_new = booking.split(',')
                                booking_new[-3] = check_in.strftime("%d/%m/%Y")
                                x.append(','.join(booking_new))
                            else:
                                x.append(booking)
                        all_booking.truncate(0)
                        
                    with open('booking.txt','w') as all_booking:
                        for i in x:
                            all_booking.writelines(i)
                            
    
                case '3':
                    while True:
                        check_out = input('Enter your check out date: ')

                        try:
                            check_out = datetime.datetime.strptime(check_out, '%d/%m/%Y')
                            with open('booking.txt', 'r') as booking:
                                all_booking = booking.readlines()
                                for booking in all_booking:
                                    if confirmation_number in booking.split(','):
                                        booking_new = booking.split(',')
                                check_in = datetime.datetime.strptime(booking_new[-3], '%d/%m/%Y')
                            if check_out.date() > check_in.date():
                                break
                            else:
                                print('your check out date should be after the check in date!')
                        except:
                            print('please enter the date in day/month/year format')

                    with open('booking.txt','r+') as all_booking:
                        x=[]
                        for booking in all_booking.readlines():
                            if confirmation_number in booking.split(','):
                                booking_new = booking.split(',')
                                booking_new[-2] = check_out.strftime("%d/%m/%Y")
                                x.append(','.join(booking_new))
                            else:
                                x.append(booking)
                        all_booking.truncate(0)
                        
                    with open('booking.txt','w') as all_booking:
                        for i in x:
                            all_booking.writelines(i)

        case '2':
            with open('booking.txt','r+') as booking_confirmation:
                x=[]
                for line in booking_confirmation.readlines():
                    if confirmation_number in line:
                        continue
                    else:
                        x.append(line)
                booking_confirmation.truncate(0)
            with open('booking.txt','w') as booking_confirmation:
                booking_confirmation.writelines(x)
                            
```
